### Lets print Data Frame of the top 20 godly players

```python
duo_df['revives_kills_ratio'] = duo_df['duo_Revives'] / duo_df['duo_Kills']
duo_df[['revives_kills_ratio' , 'duo_Revives' , 'duo_Kills']].sort_values(by=['revives_kills_ratio',], ascending=False).head()
```

### Now it seems fine😍😍😊😊
- **These are the players who have maximum revives**

```python
duo_df_not_zero = duo_df[duo_df['duo_Kills']>0]
duo_df['revives_kills_ratio'] = duo_df['duo_Revives'] / duo_df_not_zero['duo_Kills']
duo_df[['player_name','revives_kills_ratio' , 'duo_Revives' , 'duo_Kills']].sort_values(by=['duo_Revives'], ascending=False).head(20)
```

### Now lets see according to biggest revives_kills_ratio
- **These are the people who have revived more times than killed**

```python
duo_df_not_zero = duo_df[duo_df['duo_Kills']>0]
duo_df['revives_kills_ratio'] = duo_df['duo_Revives'] / duo_df_not_zero['duo_Kills']
duo_df[['player_name','revives_kills_ratio' , 'duo_Revives' , 'duo_Kills']].sort_values(by=['revives_kills_ratio','duo_Revives'], ascending=False).head(20)
```

### Now lets see according to duo_kills
- **These are the people who have killed more times than revived**

```python
duo_df_not_zero = duo_df[duo_df['duo_Kills']>0]
duo_df['revives_kills_ratio'] = duo_df['duo_Revives'] / duo_df_not_zero['duo_Kills']
duo_df[['player_name','revives_kills_ratio' , 'duo_Revives' , 'duo_Kills']].sort_values(by=['duo_Kills'], ascending=False).head(20)
```
