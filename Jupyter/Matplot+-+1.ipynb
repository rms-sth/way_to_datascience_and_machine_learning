```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(1,5,11)
```

```python
y = x **2
```

# Two ways to create plot

## 1) Functional Method

```python
plt.plot(x,y)
```

## Actual Output is from plt.show()

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('First Graph')
```

# Multiplot in same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
```

```python
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# 2) Object Oreinted Method:

```python
fig = plt.figure() #creating an empty figure i.e. empty canvas
```

```python
my_axes = fig.add_axes([0.1,0.1,0.8,0.9]) #Doesnt give result if [17] cell and [39] cell are separated
```

```python
fig = plt.figure()
my_axes = fig.add_axes([0.1,0.1,0.8,0.8])
my_axes.plot(x,y)
my_axes.set_xlabel('X-Axis')
my_axes.set_ylabel('Y-Axis')
my_axes.set_title('Simple Graph')
```

# Graph within graph

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.2]) # 20% from left axes, 50% from bottom, 30% of canvas size in width, 20% of canvas size in height
axes3 = fig.add_axes([0.45,0.2,0.4,0.2])
axes1.plot(x,y)
axes2.plot(y,x)
axes3.plot(x**2,y)

axes1.set_title('Largest Graph')
axes2.set_title('Smallest Top Graph')
axes3.set_title('Smallest Bottom Graph')
```
