```
#Read 'filename' and store it in a dictionary called 'Dictionary'
def LoadDictionary(filename):
	Dict = {}
	fDict = open(filename, 'r')
	for line in fDict.xreadlines():
		line = line.split()
		Dict[line[0]] = True
	fDict.close()
	return Dict
```

```
def Store(dictionary, clist, cand, action, pos):
	#Check that the candidate occurs in the dictionary
	if dictionary.get(cand,False) != False:
		#Store the candidate, the applied action and the position in the candidate list
		clist.append([cand, action, pos])
```

```
def Deletion(dictionary, clist, chars, misspelled, pos):
	#Any letter can be added in any position but no space...
	for letter in chars[:-1]:
		#Create the candidate correction
		candidate = misspelled[:pos] + letter + misspelled[pos:]
		#Store the candidate if it occurs in the dictionary
		Store(dictionary, clist, candidate, 'Deletion', pos)
			

def Insertion(dictionary, clist, misspelled, pos):
	#Create the candidate correction
	candidate = misspelled[:pos] + misspelled[pos+1:]
	#Store the candidate if it occurs in the dictionary
	
	Store(dictionary, clist, candidate, 'Insertion', pos)

def Substitution(dictionary, clist, chars, misspelled, pos):
	#Any letter can be added in any position but no space...
	for letter in chars[:-1]:
		#Create the candidate correction
		candidate = misspelled[:pos] + letter + misspelled[pos+1:]
		#Store the candidate if it occurs in the dictionary
		Store(dictionary, clist, candidate, 'Substitution', pos)

def Transposition(dictionary, clist, misspelled, pos):
	#Check that there is a next letter
	if pos < len(misspelled) - 1:
		#Create the candidate correction
		candidate = misspelled[:pos] + misspelled[pos+1] + misspelled[pos] + misspelled[pos+2:]
		#Store the candidate if it occurs in the dictionary
		Store(dictionary, clist, candidate, 'Transposition', pos)
```
