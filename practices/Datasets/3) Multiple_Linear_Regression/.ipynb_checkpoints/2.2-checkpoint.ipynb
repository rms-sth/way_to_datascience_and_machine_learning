# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train) #feeding to model
y_pred = regressor.predict(X_test) #predicting for X_test
y_pred
```

```python
y_test
```

```python
pd.DataFrame(data=list(zip(y_test, y_pred)),columns=['y','y-hat'])
```

```python
regressor.score(pd.DataFrame(X_test), pd.DataFrame(y_test))
```

# Model Selection:

### 1) All in

### 2) Backward Elimination

### 3) Forward Elimination

### 4) Bidirectional Cross Elimination

### 5) Score Comparison

## Backward Elimination

#### i) Select a significance level (alpha = 0.05)

#### ii) Fit the model with all predictors (R&D)

#### iii) Consider the predictors with hightest P-values

### If P-value > Significance level, goto step iv) [If P-value of all columns is less than SL value then stop]

#### iv) Remove the predictor

#### v) Fit the model

## Forward Elimination

#### Just opposite of backward elimination. i.e take a predictors with lowest P-value and continue the process

## Bidirectional Cross Elimination

#### 1) Select the significanc level to enter and to staya in the model (SL enter = 0.05 and SL state = 0.05)

#### 2) If P is less than SL enter apply forward slection

```python
import statsmodels.formula.api as sm
import numpy as np
```

```python
np.ones([50,1]).astype(int)
```

```python
df.head()
```

```python
X = df.drop('Profit', axis=1)
X.head()
```

```python
X = np.append(arr=np.ones([50,1]).astype(int), values = X , axis =1)
# convertin into y = b0 + b1x1 + b2x2 + b3x3 + ...
# y = b0 + b1x1
```

```python
X
```

```python
X_opt = X[:,[0,1,2,3,4,5]]
regression_ols = sm.OLS(endog=df['Profit'], exog = X_opt).fit() #Ordinary Least Square
regression_ols.summary()
```

```python
X_opt= X[:,[0,1,2,3,5]]
regression_ols = sm.OLS(endog=df['Profit'], exog = X_opt).fit() #Ordinary Least Square
regression_ols.summary()
```

```python
X_opt = X[:,[0,1,2,3]]
regression_ols = sm.OLS(endog=df['Profit'], exog = X_opt).fit() #Ordinary Least Square
regression_ols.summary()
```

```python
X_opt = X[:,[0,1,3]]
regression_ols = sm.OLS(endog=df['Profit'], exog = X_opt).fit() #Ordinary Least Square
regression_ols.summary()
```

```python
X_opt = X[:,[0,1]]
regression_ols = sm.OLS(endog=df['Profit'], exog = X_opt).fit() #Ordinary Least Square
regression_ols.summary()
```
