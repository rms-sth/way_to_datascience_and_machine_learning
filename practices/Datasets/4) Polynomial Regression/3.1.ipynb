```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
```

```python
df =pd.DataFrame(boston.data)
```

```python
df.coloumns = boston.feature_names
```

```python
df['Price'] = boston.target
```

```python
df.head(8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Price', axis=1),df['Price'],test_size=0.3, random_state = 5)
```
