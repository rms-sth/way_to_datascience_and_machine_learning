```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
boston
```

```python
df = pd.DataFrame(boston.data)
df.head()
```

```python
df.columns = boston.feature_names
df.head()
```

```python
df['Price'] = boston.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.count()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Price', axis=1),df['Price'],test_size=0.3, random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_test
```

```python
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
sc_y = StandardScaler()
# y_train = sc_y.fit_transform(pd.DataFrame(y_train)) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
y_pred
```

```python
pd.DataFrame(data=list(zip(y_test, y_pred)), columns=['y_test', 'y_pred']).head(10)
```
