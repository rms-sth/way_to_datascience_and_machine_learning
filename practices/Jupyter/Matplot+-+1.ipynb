# 2) Object Oreinted Method:

```python
fig = plt.figure() #creating an empty figure i.e. empty canvas
```

```python
my_axes = fig.add_axes([0.1,0.1,0.8,0.9]) #Doesnt give result if [17] cell and [39] cell are separated
```

```python
fig = plt.figure()
my_axes = fig.add_axes([0.1,0.1,0.8,0.8])
my_axes.plot(x,y)
my_axes.set_xlabel('X-Axis')
my_axes.set_ylabel('Y-Axis')
my_axes.set_title('Simple Graph')
```

# Graph within graph

#### Moreover it is the manual procedure of creating the graph

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.2]) # 20% from left axes, 50% from bottom, 30% of canvas size in width, 20% of canvas size in height
axes3 = fig.add_axes([0.45,0.2,0.4,0.2])
axes1.plot(x,y)
axes2.plot(y,x)
axes3.plot(x**2,y)

axes1.set_title('Largest Graph')
axes2.set_title('Smallest Top Graph')
axes3.set_title('Smallest Bottom Graph')
```

## Creating Sub-plot using Object Oriented method

#### In sub-plot we can specify the number of rows and columns 
#### It is the automatic version of plt.figure()

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) 
#plt.subplots() is the axes manager on top of plt.figure().
#Also plt.subplots() automatically doing fig.add_axes([]) as above based upon the no. of rows and columns defined
```

```python
fig, axes = plt.subplots(nrows=2,ncols=3)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #we get the array of axes here
axes[0].plot(x,y)
axes[0].set_title("First Graph")
axes[1].plot(y,x)
axes[1].set_title("Second Graph")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #we get the array of axes here
axes[0].plot(x,y)
axes[0].set_title("First Graph")
axes[1].plot(y,x)
axes[1].set_title("Second Graph")
plt.tight_layout()
```

# Figure size , aspect-ratio and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,2),dpi=100) #dpi is optional
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
```

```python
fig , axes = plt.subplots(nrows = 2, ncols= 1 ,figsize=(10,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Legends

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x verses Y')
ax.plot(y,x,label='Y verses X')
ax.plot(x,x*7/2,label='x verses x * 7 / 2')
ax.legend(loc=0)
#We must need ax.legend() inorder to print the labels
#loc=0 choses the best location for our labels ie. legend
```
